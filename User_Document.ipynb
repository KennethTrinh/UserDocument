# Some background

We will primarily be using libraries to assist with plotting our dataset.  A Python library is a collection of related modules. It contains bundles of code that can be used repeatedly in different programs.  These bundles of code are typically known as **functions**.


A **function** is a block of code which only runs when it is called. It can transform the data you pass in or pretty much do anything you want.

You can pass in data, known as **arguments**, into a function.  The variables in the function definition are called **parameters**.


In Python a function is defined using the `def` keyword.  If we want to give the value that our function produced, we precede it with the ```return``` statement.  For example, let's define a function to add two numbers.  

````python
def addTwoValues(value1, value2):
    return value1 + value2

addTwoValues(1,5)
````

Note that when defining and calling our functions, ```number1``` and ```number2``` are parameters, and ```1``` and ```5``` are the arguments we pass in, both of which must be wrapped in parenthesis ```()``` and separated by commas ```,```

````python
#this is a comment, which is not executed by python. 
#Comments are followed by '#' and are valid for the remaining line
#We use comments to explain what happens with our code.
#Let's see what happens when you try to add two strings!

addTwoValues('Father', ' Dr. Mozafari')
````

If we try to add a string and an integer, we will get an error.

````python
addTwoValues('Father', 1)
````

# Introduction to Numpy 

Numpy is another Python library used for working with arrays, short for "Numerical Python".  In Python we have lists that serve the purpose of arrays, but they  


1.  may be tedious to generate
2.  are relatively slow to process for large datasets

Numpy arrays are up to 50x faster and have functions to create arrays faster than regular lists.  Let's illustrate how numpy simplifies our workflow.  


#  Example Problem
Suppose we want a list of numbers *starting from 0 and ending at 15 with exactly 10 equally spaced datapoints*.  Without numpy, we can use a for-loop / list comprehension technique to achieve this along with some math. The syntax: `for i in range(10)` iterates through all integers from `0...9` inclusive.  It is important to note that the `range` function only supports integers, which is why we must use this technique for solving this problem, as 15 is not divisible by 10.

````python
regularArray = [15*i / 9 for i in range(10)]
print('Regular Array: ')
print(regularArray)
````

We can replicate this exact functionality with less cumbersome syntax in numpy.  First, we would `import` numpy with the alias `np`, just as we did for matplotlib. Then, exactly one simple function call is needed: 

````python
import numpy as np

numpyArray = np.linspace(0, 15, 10)
print('Numpy Array:')
print(numpyArray)
````

Numpy arrays have another very important property where you can perform vector operations easily. Suppose we wanted to add 1 to each element in our list.  We can do this operation with another list-comprehension to our `regularArray`.  It may be easier as a simple scalar operation to  our `numpyArray`

````python
print('Adding one to regularArray: ')
print([i + 1 for i in regularArray])
print()
print('Adding one to numpyArray: ')
print( numpyArray + 1)
````


It is refreshing to know that Matplotlib supports numpy arrays. Say we want to plot the following equation:

$$ \left| \sin(x) \right| + 5 \cdot e^{-x^{100}} \cdot \cos(x) \quad for -3 \leq x \leq 3 $$

We can use built in Numpy functions that apply the above mathematical operations to the whole array!

````python
import numpy as np
x = np.linspace(-3, 3, 100)
y = np.abs( np.sin(x) ) + 5*np.exp(-x**100) * np.cos(x)
plt.plot(x,y)
plt.show()
````

# Plotting Large Datasets

Say we want the LSTAT column as a box plot... The following code will achieve that:


````python
plt.boxplot(data['LSTAT'], vert=False, patch_artist=True, labels=['LSTAT'], boxprops=dict(facecolor='red'))
plt.xlabel('Percentage of lower status of the population')
````

But what if we wanted to do a boxplot for every column.  We *could* just list them out one by one, which would be very tedious.  Or we could take advantage of iteration in python, which requires a few lines of code.  
An explanation for the function below:



*   it takes a dataframe as input called `data`
*   it creates a figure with 7 columns and 2 rows --> as a reminder, our dataframe has 7*2 = 14 columns

*   it flattens the figure into a 1D array, so each subplot is a single element in the array


*   it iterates through an enumeration of the dataframe's columns



*   it creates a boxplot for each column in the dataframe (turns into numpy  array first)
*   it displays the boxplots of the figure



````python
def boxplot(data):
    fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 9))
    axs = axs.flatten()
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
    for i, (label, numericalValues) in enumerate(data.items()):
        axs[i].boxplot( numericalValues.to_numpy(), patch_artist=True)
        axs[i].set_title(label)
    plt.show()
  
boxplot(data)
````

We can use a similar construct for displaying the histograms of certain columns, specifying the color of each histogram

````python
def hist_plot(data):
    fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 9))
    axs = axs.flatten()
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
    column_sels = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']
    for (i,k), color in zip(enumerate(column_sels), ['r', 'g', 'b', 'y', 'c', 'm', 'k', 'pink']):
        axs[i].hist(data[k].to_numpy(), color=color)
        axs[i].set_xlabel(k)
    plt.show()
  
hist_plot(data)
````

Lastly, let's compare other columns to the `MEDV` column using a scatter plot (same idea):

````python
def scatter_plot(data):
    fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 9))
    axs = axs.flatten()
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
    column_sels = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']
    for (i,k), color in zip(enumerate(column_sels), ['r', 'g', 'b', 'y', 'c', 'm', 'k', 'pink']):
        axs[i].scatter(data[k].to_numpy(), data['MEDV'].to_numpy(), color=color)
        axs[i].set_xlabel(k)
        axs[i].set_ylabel('MEDV')
    plt.show()

scatter_plot(data)
````
